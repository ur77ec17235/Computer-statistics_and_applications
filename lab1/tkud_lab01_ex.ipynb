````python
3. Viết chương trình nhập vào một dãy các số nguyên được cách nhau bởi khoảng trắng (VD: 32 8 4 6 9...) lưu vào python ở dạng mảng. Thực hiện

- Xuất mảng
- Tính tổng các phần tử của mảng
- Tìm số lớn nhất
- Tìm số bé nhất
- Sắp xếp mảng theo thứ tự giảm dần
- In ra các số chẵn trong mảng
````

````python


a = list(map(int,input().split()))

print(a)
tong = 0
Max = 0
Min = a[0]
a_down = []
a_chan = []
#tổng
#tìm số lớn nhất
#tìm số bé nhẩt
for x in a:
    tong +=x
    if x> Max :
        Max = x
    if x < Min:
        Min = x
    if x % 2 ==0:
        a_chan.append(x)

#tìm số chẵn

#sắp theo giảm dần
# Bubble sort in Python

def bubble_Sort(a_down):
    
  # loop to access each array element
  for i in range(len(a_down)):

    # loop to compare array elements
    for j in range(0, len(a_down) - i - 1):

      # compare two adjacent elements
      # change > to < to sort in descending order
      if a_down[j] > a_down[j + 1]:

        # swapping elements if elements
        # are not in the intended order
        temp = a_down[j]
        a_down[j] = a_down[j+1]
        a_down[j+1] = temp


a_down = a.copy()

bubble_Sort(a_down)


print(tong)
print(Max)
print(Min)
print(a_down)
print(a_chan)
````
