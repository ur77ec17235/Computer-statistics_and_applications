**Bài 01:** Một bài thi trắc nghiệm gồm 10 câu hỏi, mỗi câu có 4 phương án trả lời trong đó chỉ có một phương án đúng. Bạn hãy: 

a. Một sinh viên không học bài làm bài bằng cách chọn ngẫu nhiên một phương án cho mỗi câu hỏi. Bạn hãy dùng hàm random để in ra số câu đúng của bạn sinh viên trong 10 lần kiểm tra.

b. Giả sử mỗi câu đúng được 4 điểm, mỗi câu sai bị trừ 2 điểm. Tính xác suất để sinh viên này được 4 điểm

c. Gọi X là số câu trả lời đúng. Tính E(X) và Var(X)

A/ Một sinh viên không học làm bài bằng cách chọn ngẫu nhiên một phương án cho  mỗi câu hỏi. BẠn hãy dùng hàm random, để in ra số câu đúng của bạn sinh viên trong 10 lần kiểm tra

```python
n, p = 10, 0.25
```

```python
results = binom.rvs(n,p, size = 10)
print("Số câu đúng trong mỗi lần kiểm tra: ", results)
```

```python
# B. Giải sử một câu đúng được 4 điểm, mỗi câu sai bị trừ 2 điểm.
# Tính xác suất để sinh viên này được 4 điểm.

# wrong_answer = n - Correct_answers
# total_score = 4 = Correct_answers * 4 - 2 * wrong_answers
correct_answers_requested = (4 + n *2)/6
if correct_answers_requested.is_integer() == True:
    correct_answers_requested = int(correct_answers_requested)
    
    probability= binom.pmf(correct_answers_requested, n, p)
    print(f"Xác suất để sinh viên được 4 điểm là: {probability:.4f}.")
else:
    print("Không có số câu đúng nguyên nào để đạt được 4 điểm, xác suất là 0.")
```

```python
# Goị X là số câu trả lời đúng. Tính E(X) và Var(X)

# Kì vọng của X (E(X))
Ex = n * p
#  Giá trị của X
VarX = Ex *(1-p)

print("E(x) là ",Ex)
print("Var(x) là ",VarX)
```

**Bài 02:** Một trung tâm bưu điện nhận được 3 cuộc gọi điện thoại mỗi phút. Tính xác suất để trung tâm nhận được 1 cuộc gọi, 2 cuộc gọi, 3 cuộc gọi trong vòng một phút biết số cuộc gọi trong một phút có phân phối Poisson.

```python
lambda_call = 3

p_1_call = poisson.pmf(1,lambda_call)
p_2_call = poisson.pmf(2,lambda_call)
p_3_call = poisson.pmf(3,lambda_call)

```

```python
print(f'Xác suất để nhận được một cuộc gọi là {p_1_call:.4f}')
print(f'Xác suất để nhận được hai cuộc gọi là {p_2_call:.4f}')
print(f'Xác suất để nhận được ba cuộc gọi là {p_3_call:.4f}')
```

Loại 1: trên 505 gam

Loại 2: từ 495 – 505 gam

Loại 3: dười 495 gam

c. Bạn hãy mô phỏng lấy 100 trái cây trên. Bạn hãy vẽ histogram trọng lượng trái cây và tính giá trị trung bình và độ lệch chuẩn trái cây từ mẫu

```python
#hãy tính tỉ lệ  mỗi loại 


sigma = 4
mu   =  500
x1 = 505
x3 = 495


p_type_1 = 1 - norm.cdf(x1, mu, sigma)#với loại 1 : trên 505
p_type_3 = norm.cdf(x3, mu, sigma) #với loại 2: từ 495 - 505
p_type_2 = norm.cdf(x1, mu, sigma) - norm.cdf(x3, mu, sigma) #với loại 3 dưới 495

print(f"Trái cây thu được theo loại 1 là: {p_type_1* 100:.2f}%")
print(f"Trái cây thu được theo loại 2 là: {p_type_2* 100:.2f}%")
print(f"Trái cây thu được theo loại 3 là: {p_type_3* 100:.2f}%")
```

```python
#b. hãy thử mô phỏng lấy mẫu 20 trái cây trên. Bnạ hãy vẽ histogram trọng
#lượng trái cây và tính gái trị trung bình và độ lệch chuẩn trái cây từ mẫu 

# Lấy mẫu 20 trái cây
samples_20 = np.random.normal(mu, sigma, 20)

# Vẽ histogram
plt.hist(samples_20, bins=5, edgecolor='black')
plt.title('Histogram of 20 Sampled Fruits')
plt.xlabel('Weight (g)')
plt.ylabel('Frequency')
plt.show()

# Tính trung bình và độ lệch chuẩn từ mẫu
mean_20 = np.mean(samples_20)
std_20 = np.std(samples_20)

print(f'Trung bình mẫu 20 trái cây: {mean_20:.2f} gam')
print(f'Độ lệch chuẩn mẫu 20 trái cây: {std_20:.2f} gam')
```

```python
# Lấy mẫu 100 trái cây
samples_100 = np.random.normal(mu, sigma, 100)

# Vẽ histogram
plt.hist(samples_100, bins=10, edgecolor='black')
plt.title('Histogram of 100 Sampled Fruits')
plt.xlabel('Weight (g)')
plt.ylabel('Frequency')
plt.show()

# Tính trung bình và độ lệch chuẩn từ mẫu
mean_100 = np.mean(samples_100)
std_100 = np.std(samples_100)

print(f'Trung bình mẫu 100 trái cây: {mean_100:.2f} gam')
print(f'Độ lệch chuẩn mẫu 100 trái cây: {std_100:.2f} gam')
```

**Bài 04:** Mô phỏng tung một con xúc sắc cân đối đồng chất 5000 lần. Dựa vào giá trị mô phỏng, bạn hãy tìm các giá trị xác suất dưới đây:

a. Xác suất để  số chấm xuất hiện là 4

b. Xác suất để số chấm xuất hiện lớn hơn hoặc bằng 4

c. Giả sử biết số chấm xuất hiện lớn hơn hoặc bằng 4. Hãy tìm xác suất để mặt 6 chấm xuất hiện.

```python


np.random.seed(0)  
n_rolls = 5000
rolls = np.random.randint(1, 7, n_rolls)  # Xúc xắc có giá trị từ 1 đến 6

# a. Xác suất để số chấm xuất hiện là 4
p_is_4 = np.mean(rolls == 4)

# b. Xác suất để số chấm xuất hiện lớn hơn hoặc bằng 4
p_gte_4 = np.mean(rolls >= 4)

# c. Xác suất để mặt 6 chấm xuất hiện, biết số chấm lớn hơn hoặc bằng 4
p_6_given_gte_4 = np.mean(rolls[rolls >= 4] == 6)

print("Xác suất để số chấm xuất hiện là 4 là ", p_is_4)
print("xác suất để số chấm xuất hiện lớn hơn hoặc bằng 4 là ", p_gte_4)
print("Xác suất để mặt 6 chấm xuất hiện ", p_6_given_gte_4)


```
