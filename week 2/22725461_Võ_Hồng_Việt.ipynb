````python


a = list(map(int,input().split()))

print(a)
tong = 0
Max = 0
Min = a[0]
a_down = []
a_chan = []
#tổng
#tìm số lớn nhất
#tìm số bé nhẩt
for x in a:
    tong +=x
    if x> Max :
        Max = x
    if x < Min:
        Min = x
    if x % 2 ==0:
        a_chan.append(x)

#tìm số chẵn

#sắp theo giảm dần
# Bubble sort in Python

def bubble_Sort(a_down):
    
  # loop to access each array element
  for i in range(len(a_down)):

    # loop to compare array elements
    for j in range(0, len(a_down) - i - 1):

      # compare two adjacent elements
      # change > to < to sort in descending order
      if a_down[j] > a_down[j + 1]:

        # swapping elements if elements
        # are not in the intended order
        temp = a_down[j]
        a_down[j] = a_down[j+1]
        a_down[j+1] = temp


a_down = a.copy()

bubble_Sort(a_down)


print(tong)
print(Max)
print(Min)
print(a_down)
print(a_chan)
````

4. Viết chương trình yêu cầu người dùng nhập họ tên. Hiển thị tên người dùng sau khi đã chuẩn hóa. Biết rằng tên đã chuẩn hóa là:
- Ký tự đầu viết hoa
- Không có khoảng trắng thừa
- Không có ký tự đặc biệt

````python
import re

def chuan_hoa_ten(ten):
    ten = re.sub(r'[^a-zA-Z\s]', '', ten)
    ten = ten.title() 
    # Loại bỏ khoảng trắng thừa ở đầu, cuối và giữa các từ
    ten = ten.strip()  
    ten = re.sub(r'\s+', '', ten)  # Thay thế khoảng trắng thừa giữa các từ bằng một khoảng trắng duy nhất
    
    
    return ten


ho_ten = input("Nhập họ và tên: ")

ten_chuan_hoa = chuan_hoa_ten(ho_ten)

print(f"Tên đã chuẩn hóa: {ten_chuan_hoa}")
````

5. Viết chương trình nhập vào một số nguyên dương (n>0). Cho biết n có phải là số đối xứng hay không? Biết số đối xứng là số viết xuôi hay viết ngược đều như nhau

````python


# the function to find the given number is mirror?
def is_palindrome(number):
    # Chuyển số thành chuỗi
    str_number = str(number)
    
    # So sánh các ký tự đầu và cuối
    for i in range(len(str_number) // 2):
        if str_number[i] != str_number[-i - 1]:
            return False
    return True

# Kiểm tra số đối xứng
num = int(input("Nhập số cần kiểm tra: "))
if is_palindrome(num):
    print(f"{num} là số đối xứng")
else:
    print(f"{num} không phải là số đối xứng")
````

8. Viết chương trình chấp nhận một chuỗi số, phân tách bằng dấu phẩy từ giao diện điều khiển, tạo ra một List và một tuple chứa mọi số.

    Ví dụ:

    ```
    Đầu vào được cung cấp là '34,67,55,33,12,98' thì đầu ra là:

    [34, 67, 55, 33, 12, 98]
    (34, 67, 55, 33, 12, 98)
    ```


````python
# Yêu cầu người dùng nhập chuỗi số, phân tách bằng dấu phẩy
input_string = input("Nhập chuỗi số, phân tách bằng dấu phẩy: ")

# Tạo danh sách bằng cách tách chuỗi dựa trên dấu phẩy
list_numbers = input_string.split(',')

# Chuyển đổi từng phần tử trong danh sách thành số nguyên
list_numbers = [int(i) for i in list_numbers]

# Tạo tuple từ danh sách
tuple_numbers = tuple(list_numbers)

# Hiển thị danh sách và tuple
print(list_numbers)
print(tuple_numbers)
````

9. Viết hàm nhận vào một danh sách các số nguyên. Hãy viết hàm trả về một danh sách gồm các số không trùng nhau trong danh sách ban đầu

````python
def unique_numbers(lst):
    # Sử dụng set để loại bỏ các phần tử trùng lặp, sau đó chuyển về danh sách
    return list(set(lst))

# Ví dụ sử dụng hàm
input_list = list(input())
result = unique_numbers(input_list)

print(f"các số không trùng nhau là {result}")
````

10. Viết hàm có tham số đầu vào là một list L có các phần tử là chuỗi. Hãy tạo ra một dictionary D mã hóa, với mỗi một phần tử trong L được mã hóa thành một con số (theo thứ tự từ 0 tăng dần lên 1 đơn vị). Sau đó trả về list đã được mã hóa

    Ví dụ:

    ```
    Cho L = ["đen","vàng","xanh","vàng","xanh","đỏ","hồng"]

    Xây dựng dictionary mã hóa:
    D = {'đen':0,'vàng':1,'xanh':2,'đỏ':3,'hồng':4}

    Trả về List mã hóa:
    L_mahoa = [0, 1, 2, 1, 2, 3, 4]
    ```

````python
def ma_hoa_list(input_string):
    # Tách chuỗi đầu vào thành danh sách các phần tử
    L = input_string.split(',')

    # Tạo dictionary mã hóa với giá trị tăng dần từ 0
    D = {}
    ma_hoa = 0
    for item in L:
        item = item.strip()  # Loại bỏ khoảng trắng thừa
        if item not in D:
            D[item] = ma_hoa
            ma_hoa += 1

    # Mã hóa danh sách dựa trên dictionary đã tạo
    L_mahoa = [D[item.strip()] for item in L]

    return D, L_mahoa

# Ví dụ sử dụng hàm
input_string = str(input())
D, L_mahoa = ma_hoa_list(input_string)

# Hiển thị dictionary mã hóa và list đã mã hóa
print("Dictionary mã hóa:", D)
print("List đã mã hóa:", L_mahoa)
````
