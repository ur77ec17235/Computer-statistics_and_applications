Bạn hãy phát sinh ngẫu nhiên mảng ndarray có tên `arr` gồm 100 phần tử số nguyên có giá trị từ 0..1000. Thực hiện các thao tác sau:

```python
import math
import random
from random import randint
import numpy as np
```

```python
arr = []
for x in range(100):
    arr.append(randint(0, 1000))

print(arr)
```

### 1. Hiển thị các phần tử thỏa yêu cầu dưới đây

- 10 phần tử đầu tiên
- 5 phần tử cuối cùng
- Các phần tử thỏa mãn điều kiện chia hết cho 5 và lớn hơn 100

```python
print("10 phần tử đầu tiên là gồm:")
for x in range(10):
    print( arr[x], end = ", ")
print("\n 5 phần tử cuối cùng là gồm:")
for x in range(len(arr)-5, len(arr)):
    print(arr[x], end = ", ")

print("\n các phần tử chia hết cho 5 và lớn hơn 100 là:")

for x in arr:
    if (x % 5 == 0) & (x > 100) :
        print(x, end = ", " )
```

### 2. Cho biết mảng trên có bao nhiêu phần tử có giá trị nhỏ hơn 100. Hiển thị các phần tử đó

```python
num =0

for x in arr:
    if x < 100:
        num +=1

print(f" số phần từ nhỏ hơn 100 là {num}")
```

### 3. Thực hiện thống kê:

- Tính tổng mảng
- Tìm giá trị trung bình
- Tìm giá trị trung vị
- Tìm Q1, Q2, Q3
- Tìm phân vị 20 (percentile 20)

```python
# sort the array

for i in range(len(arr) - 1):
    for j in range(i+1, len(arr)):
        if arr[i] > arr[j]:
            temp = arr[j]
            arr[j] = arr[i]
            arr[i] = temp
print(arr)
```

```python
sum = 0

for x in arr:
    sum +=x

print(f"giá trị tổng của các phần tử trong mảng là {sum}")

trungbinh = sum/ len(arr)
print(f"giá trị trung bình của các phần tử trong mảng là {trungbinh}")


x2_index = int((len(arr)+1)/2)
x2 = arr[x2_index]
y2 = arr[x2_index+1]
trungvi = ((x2+y2)/2)

print(f"giá trị trung vị của các phần tử trong mảng là {trungvi}")

Q2 = (trungvi)

x1_index = int((len(arr)+1)/4)
x1 = arr[x1_index]
y1 = arr[x1_index+1]
Q1 = ((x1+y1)/2)

x3_index = int((len(arr)+1)*3/4)
x3 = arr[x3_index]
y3 = arr[x3_index+1]
Q3 = ((x3+y3)/2)

print(f"giá trị Q1, Q2, Q3 lần lượt là: {Q1, Q2, Q3}")

# for x in range(20):
#     print( arr[x], end = ", ")

per20 = arr[int(20/100 * (len(arr)+1))]
print(f"giá trị phân vị 20 của các phần tử trong mảng là {per20}")
```

### 5. Tìm phần tử xuất hiện nhiều nhất trong mảng

```python
max = 1
ts = 0

for x in range(len(arr)-1):
    if arr[x] == arr[x+1]:
        max += 1
        if max > ts:
            ts = max
            a_max = arr[x]
    else:
        max = 1
print("số xuất hiện nhiều nhất là", a_max, "với số lần là: " , ts)
```

### Câu 4: Định dạng đường thẳng

Phát sinh x là các điểm dữ liệu từ $[0, \pi]$. Bạn hãy vẽ đồ thị $y = sin(x)$ và $y = cos(x)$ như dưới đây.
Gợi ý: sử dụng hằng số pi và hàm sau:
- `np.pi` : số $\pi$
- `np.linspace()`
- `np.sin()`
- `np.cos()`

```python
# This cell contains output. Don't remove!
```

```python
# YOUR CODE HERE

# Generate data for the sine wave
x = np.linspace(0, 2 * np.pi, 1000)
x_sin = np.sin(x)


y = np.linspace(0, 2 * np.pi, 1000)
y_cos = np.cos(y)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, x_sin, label='Sin(x)')
plt.plot(y, y_cos, label = "Cos(y)", ls = "--")
# Add labels and title
plt.title("Đồ thị hàm số Sin(x) và Cos(y)")

plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```
